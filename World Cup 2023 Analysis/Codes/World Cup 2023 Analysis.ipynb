## Exploratory Data Analysis (EDA)

### Seeing the Columns of Dataset

1. **team**: This column indicates the name of the cricket team participating in the World Cup 2023. Each row corresponds to a specific team.

2. **player**: This column contains the names of individual players who participated in matches during the tournament. Each row corresponds to a different player.

3. **bat_or_bowl**: This column specifies whether the player was involved in batting or bowling during the match. It could have values like "bat" for batting or "bowl" for bowling.

4. **bb_bf**: This column represents the count of balls faced by a batsman (if the player is batting) or bowled by a bowler (if the player is bowling). It provides insight into the involvement of the player during the match.

5. **runs**: This column indicates the total number of runs scored by the batsman (if the player is batting) or conceded by the bowler (if the player is bowling) during the match.

6. **wkts**: This column signifies the total number of wickets taken by the bowler during the match. It measures the effectiveness of the bowler in taking wickets.

7. **wicketball_prob**: This column might represent the probability of taking a wicket with a particular ball bowled by the player. It could provide insights into the bowling strategy and effectiveness of the player.

8. **runs_per_ball**: This column calculates the average number of runs scored (if the player is batting) or conceded (if the player is bowling) per ball. It provides a measure of the player's efficiency or effectiveness in scoring or preventing runs.

9. **opposition**: This column specifies the opposing team against which the player performed during the match. It provides context regarding the level of competition faced by the player.

10. **ground**: This column denotes the venue or stadium where the match took place. It provides information about the location of the match.

11. **start_date**: This column records the date on which the match occurred. It provides a temporal reference for the match.

12. **overs**: This column indicates the total number of overs bowled by the bowler (if the player is bowling) or faced by the batsman (if the player is batting). It measures the workload or duration of the player's involvement in the match.

13. **mdns**: This column represents the number of maiden overs bowled by the bowler. A maiden over is an over in which no runs are scored by the batting team. It reflects the bowler's ability to maintain pressure on the batsmen.

14. **econ**: This column calculates the economy rate of the bowler, indicating the average number of runs conceded per over bowled. It provides a measure of the bowler's efficiency in restricting runs.

15. **inns**: This column signifies the innings number in which the player performed during the match. It could take values like 1 or 2, indicating whether the player performed in the first or second innings of the match, respectively.

16. **4s**: This column represents the number of boundaries (fours) hit by the batsman during the match. It measures the batsman's ability to find gaps in the field and score runs.

17. **6s**: This column represents the number of sixes hit by the batsman during the match. It indicates the batsman's ability to clear the boundary ropes and score maximum runs with a single stroke.

18. **sr**: This column denotes the strike rate of the batsman. It calculates the average number of runs scored per 100 balls faced by the batsman. It provides a measure of the batsman's scoring rate and aggressiveness.

19. **not_out**: This column indicates whether the batsman remained not out during their innings. It could have values like "yes" or "no", indicating whether the batsman was dismissed or not at the end of their innings.

20. **mins**: This column represents the duration of the innings played by the batsman or bowler in minutes. It measures the time spent by the player on the field during their innings or bowling spell.

## 1. Team Performance Analysis

- Explore team-wise performance metrics. 
- Analyze runs scored, wickets taken, and batting/bowling styles. 
- Identify top-performing teams and their strengths. 

### 1.1 Explore team-wise performance metrics

```python
# Team-wise performance metrics
team_performance = df.groupby('team').agg({
    'runs': 'sum',
    'wkts': 'sum',
    'bb_bf': 'sum',
    '4s': 'sum',
    '6s': 'sum',
    'not_out': 'sum',
})


team_performance['batting_average'] = team_performance['runs'] / (team_performance['not_out'] + 1)
team_performance['bowling_average'] = team_performance['runs'] / team_performance['wkts']
team_performance['strike_rate'] = (team_performance['4s'] * 4 + team_performance['6s'] * 6) / team_performance['bb_bf']

# Plotting
plt.figure(figsize=(12, 8))

# Batting Average
plt.subplot(2, 2, 1)
team_performance['batting_average'].plot(kind='bar', color='skyblue')
plt.title('Batting Average')
plt.xlabel('Team')
plt.ylabel('Average')

# Bowling Average
plt.subplot(2, 2, 2)
team_performance['bowling_average'].plot(kind='bar', color='lightgreen')
plt.title('Bowling Average')
plt.xlabel('Team')
plt.ylabel('Average')

# Strike Rate
plt.subplot(2, 2, 3)
team_performance['strike_rate'].plot(kind='bar', color='salmon')
plt.title('Strike Rate')
plt.xlabel('Team')
plt.ylabel('Rate')

plt.tight_layout()
plt.show()
```

### 1.1.1 **Bar Graph of Total Runs Scored by Each Team:**
   - Description: This graph displays the total runs scored by each team throughout the tournament, providing an overview of batting performance.
   - Graph Type: Bar Graph
   - X-axis: Team Names
   - Y-axis: Total Runs Scored
   - Each bar represents a different team, and the height of the bar represents the total runs scored by that team.

```python
# Grouping the dataset by team and summing up the runs scored
team_runs = df.groupby('team')['runs'].sum().reset_index()

# Creating the bar plot
plt.figure(figsize=(10, 6))
plt.bar(team_runs['team'], team_runs['runs'], color='skyblue')
plt.xlabel('Team')
plt.ylabel('Total Runs Scored')
plt.title('Total Runs Scored by Each Team')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 1.1.2 **Bar Graph of Total Wickets Taken by Each Team:**
   - Description: This graph illustrates the total number of wickets taken by each team, indicating their bowling performance.
   - Graph Type: Bar Graph
   - X-axis: Team Names
   - Y-axis: Total Wickets Taken
   - Each bar represents a different team, and the height of the bar represents the total number of wickets taken by that team.

```python
# Grouping the dataset by team and summing up the wickets taken
team_wickets = df.groupby('team')['wkts'].sum().reset_index()

# Creating the bar plot
plt.figure(figsize=(10, 6))
plt.bar(team_wickets['team'], team_wickets['wkts'], color='lightgreen')
plt.xlabel('Team')
plt.ylabel('Total Wickets Taken')
plt.title('Total Wickets Taken by Each Team')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

 ### 1.2 Analyze runs scored, wickets taken, and batting/bowling styles

### 1.2.1 **Stacked Bar Graph of Runs Scored by Batting Style for Each Team:**
   - Description: This graph compares the runs scored by each team using different batting styles, such as boundaries (4s and 6s) and singles.
   - Graph Type: Stacked Bar Graph
   - X-axis: Team Names
   - Y-axis: Total Runs Scored
   - Each bar represents a different team, and each segment of the bar represents the runs scored through a specific batting style (e.g., 4s, 6s, singles).

```python
# Grouping the dataset by team and summing up the runs scored through different batting styles
team_batting_style = df.groupby('team')[['4s', '6s', 'runs']].sum().reset_index()

# Creating the stacked bar plot
plt.figure(figsize=(10, 6))
plt.bar(team_batting_style['team'], team_batting_style['4s'], label='4s', color='lightblue')
plt.bar(team_batting_style['team'], team_batting_style['6s'], bottom=team_batting_style['4s'], label='6s', color='blue')
plt.bar(team_batting_style['team'], team_batting_style['runs'] - team_batting_style['4s'] - team_batting_style['6s'],
        bottom=team_batting_style['4s'] + team_batting_style['6s'], label='Singles', color='cyan')
plt.xlabel('Team')
plt.ylabel('Total Runs Scored')
plt.title('Runs Scored by Batting Style for Each Team')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

### 1.2.2 **Stacked Bar Graph of Wickets Taken by Bowling Style for Each Team:**
   - Description: This graph compares the wickets taken by each team using different bowling styles, such as regular wickets or caught behind.
   - Graph Type: Stacked Bar Graph
   - X-axis: Team Names
   - Y-axis: Total Wickets Taken
   - Each bar represents a different team, and each segment of the bar represents the wickets taken through a specific bowling style (e.g., regular wickets, caught behind).

```python
# Grouping the dataset by team and summing up the wickets taken through different bowling styles
team_bowling_style = df.groupby('team')[['wkts', 'mdns']].sum().reset_index()

# Creating the stacked bar plot
plt.figure(figsize=(10, 6))
plt.bar(team_bowling_style['team'], team_bowling_style['wkts'], label='Regular Wickets', color='lightgreen')
plt.bar(team_bowling_style['team'], team_bowling_style['mdns'], bottom=team_bowling_style['wkts'], label='Maiden Overs', color='green')
plt.xlabel('Team')
plt.ylabel('Total Wickets Taken')
plt.title('Wickets Taken by Bowling Style for Each Team')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

### 1.2.3 **Pie Chart of Batting/Bowling Style Distribution for Top Performing Teams:**
   - Description: This pie chart illustrates the distribution of runs scored or wickets taken by the top-performing teams, showcasing their predominant batting or bowling styles.
   - Graph Type: Pie Chart
   - Each slice of the pie represents a different batting or bowling style, and the size of the slice represents the proportion of runs scored or wickets taken through that style for the top-performing teams.

```python
# Top_teams is a list of top performing teams
top_teams = ['NZ', 'AUS', 'SA', 'PAK', 'IND']

# Calculate the total runs scored and wickets taken by batting/bowling style for top performing teams
top_teams_data = df[df['team'].isin(top_teams)]
batting_style_distribution = top_teams_data[['4s', '6s', 'runs']].sum()
bowling_style_distribution = top_teams_data[['wkts', 'mdns']].sum()

# Creating the pie chart for batting style distribution
plt.figure(figsize=(10, 6))
plt.pie(batting_style_distribution, labels=batting_style_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title('Batting Style Distribution for Top Performing Teams')
plt.axis('equal')
plt.tight_layout()
plt.show()

# Creating the pie chart for bowling style distribution
plt.figure(figsize=(10, 6))
plt.pie(bowling_style_distribution, labels=bowling_style_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title('Bowling Style Distribution for Top Performing Teams')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

### 1.3  Identify top-performing teams and their strengths

### 1.3.1 Bar graph of top performing teams

```python
# Calculate total runs scored and total wickets taken by each team
team_performance = df.groupby('team').agg({'runs': 'sum', 'wkts': 'sum'}).reset_index()

# Identify top-performing teams based on total runs scored and total wickets taken
top_teams = team_performance.nlargest(5, ['runs', 'wkts'])

# Plotting the performance of top-performing teams
plt.figure(figsize=(10, 6))

# Plotting total runs scored
plt.bar(top_teams['team'], top_teams['runs'], color='skyblue', label='Total Runs')

# Plotting total wickets taken
plt.bar(top_teams['team'], top_teams['wkts'], color='lightgreen', label='Total Wickets')

plt.xlabel('Team')
plt.ylabel('Count')
plt.title('Top Performing Teams and Their Strengths')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("Top Performing Teams:")
print(top_teams)
```

### 1.3.2 Pie chart of top performing teams

```python
import warnings
warnings.filterwarnings("ignore")

# Define top performing teams
top_teams_performance = ['NZ', 'AUS', 'SA', 'PAK', 'IND']

# Create a DataFrame for top_teams_performance
top_teams_df = team_performance[team_performance['team'].isin(top_teams_performance)]

# Calculate total runs and wickets across all teams
total_runs = team_performance['runs'].sum()
total_wickets = team_performance['wkts'].sum()

# Calculate contribution of runs and wickets for each top team
top_teams_df['runs_contribution'] = top_teams_df['runs'] / total_runs * 100
top_teams_df['wickets_contribution'] = top_teams_df['wkts'] / total_wickets * 100

# Plotting contribution of runs
plt.figure(figsize=(10, 6))
plt.pie(top_teams_df['runs_contribution'], labels=top_teams_df['team'], autopct='%1.1f%%', startangle=90)
plt.title('Contribution of Runs to Total Score for Top Performing Teams')
plt.axis('equal')
plt.tight_layout()
plt.show()

# Plotting contribution of wickets
plt.figure(figsize=(10, 6))
plt.pie(top_teams_df['wickets_contribution'], labels=top_teams_df['team'], autopct='%1.1f%%', startangle=90)
plt.title('Contribution of Wickets to Total Wickets Taken for Top Performing Teams')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

### 3. Opposition and Ground Analysis: 
- Investigate how teams and players perform against different oppositions. 
- Examine performance variations across different playing grounds. 
- Identify if there are specific teams or players that excel in certain conditions. 

### 3.1 **Team-wise Runs Scored Against Different Oppositions:**
- This plot illustrates how different cricket teams perform in terms of runs scored against various oppositions. Each line represents a different team, showing their runs scored against different oppositions. It helps in understanding the batting prowess of teams against different opponents.

```python
# Opposition and Ground Analysis
# Investigate how teams and players perform against different oppositions.
# Examine performance variations across different playing grounds.
# Identify if there are specific teams or players that excel in certain conditions.

# Team-wise performance against different oppositions
team_opposition_performance = df.groupby(['team', 'opposition']).agg({'runs': 'sum', 'wkts': 'sum'}).reset_index()

# Plotting team-wise runs scored against different oppositions
plt.figure(figsize=(12, 6))
for team in team_opposition_performance['team'].unique():
    data = team_opposition_performance[team_opposition_performance['team'] == team]
    plt.plot(data['opposition'], data['runs'], marker='o', label=team)

plt.xlabel('Opposition')
plt.ylabel('Runs')
plt.title('Team-wise Runs Scored Against Different Oppositions')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 3.2. **Team-wise Wickets Taken Against Different Oppositions:**
- Similar to the plot for runs scored, we can plot the number of wickets taken by each team against different oppositions. This will provide insights into team bowling performances against specific opponents.

```python
plt.figure(figsize=(12, 6))
for team in team_opposition_performance['team'].unique():
    data = team_opposition_performance[team_opposition_performance['team'] == team]
    plt.plot(data['opposition'], data['wkts'], marker='o', label=team)

plt.xlabel('Opposition')
plt.ylabel('Wickets')
plt.title('Team-wise Wickets Taken Against Different Oppositions')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 3.3 **Comparison of Runs and Wickets Taken by Teams at Different Grounds:**
- This visualization can help in comparing the batting and bowling performances of teams at different grounds, providing insights into which grounds favor batting or bowling.

```python
team_ground_performance = df.groupby(['team', 'ground']).agg({'runs': 'sum', 'wkts': 'sum'}).reset_index()

plt.figure(figsize=(12, 6))
for team in team_ground_performance['team'].unique():
    data = team_ground_performance[team_ground_performance['team'] == team]
    plt.plot(data['ground'], data['runs'], marker='o', label=f'{team} - Runs')
    plt.plot(data['ground'], data['wkts'], marker='o', linestyle='dashed', label=f'{team} - Wickets')

plt.xlabel('Ground')
plt.ylabel('Count')
plt.title('Team-wise Runs and Wickets Taken at Different Grounds')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 4. Temporal Analysis: 
- Study performance trends over time, considering start dates and overs played. 
- Identify any temporal patterns or changes in team and player performance

### 4.1 **Team-wise Performance Over Time Bar Graph**:

- The line chart plotted represents the team-wise performance over time, where the x-axis denotes the start date of matches, and the y-axis indicates the count of runs scored (in sky blue) and wickets taken (in light green). By plotting both runs and wickets on the same graph, it enables the comparison of batting and bowling performances over different time periods.

```python
# Temporal Analysis
# Study performance trends over time, considering start dates and overs played.
# Identify any temporal patterns or changes in team and player performance.

# Team-wise performance over time
team_time_performance = df.groupby('start_date').agg({'runs': 'sum', 'wkts': 'sum'}).reset_index()

plt.figure(figsize=(12, 6))
plt.plot(team_time_performance['start_date'], team_time_performance['runs'], label='Total Runs', color='skyblue')
plt.plot(team_time_performance['start_date'], team_time_performance['wkts'], label='Total Wickets', color='lightgreen')
plt.xlabel('Start Date')
plt.ylabel('Count')
plt.title('Team-wise Performance Over Time')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 4.2 **Line Plot of Team-wise performance over time**

```python
# Team-wise performance over time (Comparison)
plt.figure(figsize=(12, 6))
for team in df['team'].unique():
    team_data = df[df['team'] == team].groupby('start_date').agg({'runs': 'sum', 'wkts': 'sum'}).reset_index()
    plt.plot(team_data['start_date'], team_data['runs'], label=f'{team} - Runs')
    plt.plot(team_data['start_date'], team_data['wkts'], label=f'{team} - Wickets')

plt.xlabel('Start Date')
plt.ylabel('Count')
plt.title('Team-wise Performance Over Time (Comparison)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
